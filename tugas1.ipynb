GAMBAR RGB>GREYSCALE>BINER

```python
from PIL import Image
import numpy as np
import matplotlib
import matplotlib.pyplot  as  plt
%matplotlib inline

#mengambil gambar asli dari direktori 
img = Image.open (r"C:\Users\Toshiba\Documents\DATA USER\DATA ZUHRO\COLLAGE\KULIAH\SEMESTER 5\PENGOLAHAN CITRA DIGITAL\gbr1.jpg")

#panggil gambar awal (RGB) dengan ukuran 250x250 pixels
plt.imshow(img)
```

```python
img.save('gbr1rgb.jpg')
```

```python
#template citra baru (greyscale)
imgA = Image.new ('L',(250,250),(0))
for x in range (250) :
    for y in range (250) :
        warna = img.getpixel ((x,y))
        R = warna [0]
        G = warna [1]
        B = warna [2]
        gs = int ((0.2989*R)+(0.5870*G)+(0.1141*B))
        imgA.putpixel ((x,y),(gs))
plt.imshow (imgA)
```

```python
imgA.save('gbr1gs.jpg')
```

```python
batasan = 128
for x in range (250) :
    for y in range (250) :
        grey = imgA.getpixel ((x,y))
        if grey>= batasan :
            imgA.putpixel ((x,y),(256))
        else :
            imgA.putpixel ((x,y),(0))
plt.imshow (imgA)
```

```python
imgA.save('gbr1biner.jpg')
```

HISTOGRAM NORMAL DAN NORMALISASI

```python
#tampilkan histogram normal (contoh dari gambar RGB)
import numpy as np
import cv2 as cv
import matplotlib
import matplotlib.pyplot as plt

img = cv.imread ('gbr1rgb.jpg')
histogram = cv.calcHist ([img], [0], None, [256], [0,256])
plt.plot(histogram)
plt.show()

#tampilkan histogram normalisasi (contoh dari gambar RGB)
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread ('gbr1rgb.jpg',0)
hist, bins = np.histogram(img.flatten(), 256, [0,256])
cdf = hist.cumsum()
cdf_normalisasi = cdf*hist.max()/cdf.max()

plt.plot(cdf_normalisasi, color='blue')
plt.hist(img.flatten(),256,[0,256], color='red')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```

rata-rata, variansi, standar deviasi

```python
#GAMBAR RGB
img = cv2.imread ('gbr1rgb.jpg',0)
hist, bins = np.histogram(img.flatten(), 256, [0,256])
cdf = hist.cumsum()
i=0
for i in range (256):
    mean = sum (cdf*(i+1))   
print("\nRata-rata  = %0.2f" % mean)
print ("\nVariansi = %0.2f "%histogram.var())
print ("\nStandar Deviasi = %0.2f "%histogram.std())
```
